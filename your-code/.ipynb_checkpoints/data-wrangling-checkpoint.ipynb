```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('GSAF5.csv', encoding = 'iso-8859-1')
```

```python
data.head(2)
```

```python
#Es una tabla de accidentes por tipo de actividad acuatica.
# El objetivo: Queremos obtener que tipo de actividad es mas frecuente por pais, mes y año
#1 - Tenemos que limpiar las columnas que no vamos  necesitar 
#(Case Number, Location, Investigator or Source, pdf, href formula, href, case Number 1 y 2, original order, unnamed: 22 y 23)
```

```python
# Identificiar todas las columnas que tine el Dataset
data.columns
```

```python
# Eliminar las columnas que no vamos a usar y ajustar los nombres de las necesarias.
data = data.drop(['Case Number', 'Investigator or Source', 'pdf', 'href formula', 'href',
       'Case Number.1', 'Case Number.2', 'original order', 'Unnamed: 22',
       'Unnamed: 23'], axis = 1)

data = data.rename(index= str, columns = {'Sex ':'Sex'})
data.head()
```

```python
# Limpiar impurezas en la fecha.
# data['Date'] = data['Date'].str.replace("(\d)","-")
# data.head()
```

```python
# Quitar Espacios en columna Sex
data["Sex"] = data["Sex"].str.strip()
```

```python
# Insertar Other en los datos que no son M o F
data.loc[~data["Sex"].isin(["M","F"])] = "Other"
```
