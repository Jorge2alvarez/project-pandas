# Tabla de accidentes acuaticos 
La tabla contiene accidentes por tipo de actividad acuatica.<br>
### Objetivo: Obtener el top 5 países con mas accidentes acuaticos, por rango etario. 
1. Tenemos que limpiar las columnas que no vamos  necesitar 
2. (Case Number, Location, Investigator or Source, pdf, href formula, href, case Number 1 y 2, original order, unnamed: 22 y 23)

```python
# Identificiar todas las columnas que tine el Dataset
data.columns
```

```python
# Eliminar las columnas que no vamos a usar
data = data.drop(['Case Number', 'Name', 'Investigator or Source', 'pdf', 'href formula', 'href',
       'Case Number.1', 'Case Number.2', 'original order', 'Unnamed: 22',
       'Unnamed: 23'], axis = 1)
data.head()
```

```python
# Quitar espacios de los titulos de columna
data = data.rename(columns=lambda x: x.strip())
data.columns
```

```python
# Limpiar Fechas y formatear para la correcta visualizacion

# Limpiar impurezas (Espacios y palabras extras)

# Limpiar los datos de la columna Date.
data['Date'] = data['Date'].str.replace("Before ","")
data['Date'] = data['Date'].str.replace("No date, Before ","")
data['Date'] = data['Date'].str.replace("Reported ","")
data['Date'] = data['Date'].str.replace("No date, ","")

# Quitar Espacios
data["Date"] = data["Date"].str.strip()

# Cambiar el formato de Fecha DD/MM/YYYY
data['Date'] = pd.to_datetime(data.Date, errors='coerce')
data['Date'] = data['Date'].dt.strftime('%d/%m/%Y')
```

```python
# Completar Years Null con el año de la columna Date (los que existan)
#data['Year'].value_counts()

#data[data["Year"].isnumeric()]

data["Year"].loc[data["Year"].isin(["0"])] = 'null'
data["Year"].loc[data["Year"].isin(["Other"])] = 'null'
data["Year"].loc[data["Year"].isin(["other"])] = 'null'
```

```python
data['Year'] = pd.to_datetime(data.Date, errors='coerce')
data['Year'] = data['Year'].dt.strftime('%Y')
```

```python
#data['Year'].loc['2000':'2019']
mask = (data['Year'] > '2000') & (data['Year'] <= '2019')
mask
```

```python
# Quitar Espacios en valores de Sex
data["Sex"] = data["Sex"].str.strip()

# Si no es M o F asignar Other
data.loc[~data["Sex"].isin(["M","F"])] = "Other"
```

```python
# Si no es Y o N asignar Other
data.loc[~data["Fatal (Y/N)"].isin(["N","Y"])] = "Other"
```

```python
data["Age"].str.isnumeric().value_counts()
```

```python
# Si no es un valor numerico = (a tomar por culo)
data['Age'] = pd.to_numeric(data['Age'], errors='coerce', downcast='signed')
```

```python
data.head()
```

```python
#data['Country'].value_counts()
```

```python
# Crear Bins con las edades
age_labels = ['1', '2', '3', '4', '5']
bins = pd.DataFrame([pd.cut(data['Age'], 5, labels = age_labels), data['Age']])
```

```python
# Transpolar tabla
bins = bins.T
bins.columns = ['Bin', 'Age Control']
```

```python
# Asignar nombre a la columna Bin con los valores menor y mayor de cada bin
for i in age_labels:
    
    min_value = list(bins[bins['Bin'] == i].min())
    max_value = list(bins[bins['Bin'] == i].max())
    bin_range = str(int(min_value[1])) + ' - ' + str(int(max_value[1]))
    
    bins['Bin'].loc[bins["Bin"].isin([i])] = bin_range
```

```python
bins['Bin'].value_counts()
```

```python
# Unir las tablas Data y Bins para unificar los datos
data = data.join(bins)
```

```python
#
data_top5 = data.loc[data["Country"].isin(["USA", "AUSTRALIA", "Other", "SOUTH AFRICA", "PAPUA NEW GUINEA"])]
data_top5['count'] = 1
data_top5.head()
```
